```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/train.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Survived"].value_counts()
```

```python
target = "Survived"
features = ["Pclass", "Sex", "Age", "SibSp", "Parch", "Embarked", "Fare", "Cabin"]
```

```python
num_features = ["Age", "SibSp", "Parch", "Fare"]
cat_features = ["Pclass", "Sex", "Embarked", "Cabin"]
```

```python
for feature in num_features:
    sns.boxplot(y=df[feature], x=df[target])
    plt.show()
```

```python
for feature in cat_features:
    sns.countplot(x=df[feature], hue=df[target])
    plt.show()
```

```python
CabinClass = df["Cabin"].apply(lambda x:  x[0] if x is not np.nan else "None")
CabinClass.name = "CabinClass"
```

```python
sns.countplot(x=CabinClass, hue=df[target])
```

```python
sns.pairplot(df[features + [target]], hue=target)
```
